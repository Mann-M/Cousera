```python
import pandas as pd
import numpy as np
```

```python
df_can=pd.read_csv('Canada.csv')
```

```python
df_can.head()
```

```python
# set country as the index, as it will help us to plot charts easily, by refering to the country names as index values

df_can.set_index('Country',inplace=True)

# tip: the opposite of set is reset. So to reset the index, we can use df_can.reset()
```

```python
# lets the remove the name of the index
df_can.index.name=None
```

```python
# since we converted the years to string, let's declare a variable that allows us to call upon the full range of years:
# this will be usefull for plotting later on

years= list(map(str,range(1980,2014)))
years 
```

```python
import matplotlib as mlp
import matplotlib.pyplot as plt
```

<h2>Line Plots and their uses</h2>

<p>A line chart or line plot is a type of plot which displays information as a series of data points called 'markers' connected by straight line segments.<br>Use line plots when we have a continuous data set. These are best for trend based visualizations of data over a period of time.

```python
# plot a line graph of immigration from Haiti using df.plot()

#creating a data series
haiti=df_can.loc['Haiti', years]#passing the years 1980-2013 to exclude the "total" column
haiti.head()
```

```python
# ploting a line plot

haiti.plot()
```

```python
haiti.plot(kind="line")

plt.title('Immigration from Haiti')
plt.xlabel('Years')
plt.ylabel('No. of immigrants')

plt.show()
```

```python
haiti.index= haiti.index.map(int)
haiti.plot(kind="line")

plt.title('immigration from Haiti')
plt.xlabel('Years')
plt.ylabel('No. of immigrants')

#annotate the 2010 Earthquake
#syntax: plt.text(x,y,label)
plt.text(2000,6000,'2010 Earthquake')

plt.show()
```

```python
# Compare the number of immigrants from India and China from 1980-2013
# First we have to creat a dataFrame comprising India and China

df_CI= df_can.loc[['India','China'],years]
df_CI
```

```python
# Now we plot the line plot

df_CI.plot(kind='line')
```

Pandas plots the indices on the x-axis and the columns as individual lines on the y-axis. Since df_CI is a dataframe with the country as the index and yearsa the columns, we must first transpose the dataframe using transpose() method to swap the rows and the columns

```python
df_CI=df_CI.transpose()
df_CI
```

```python
df_CI.plot(kind='line')

plt.xlabel('years')
plt.ylabel('countries')
plt.title('India vs China Immigrants')

plt.show()
```

Line plot is a handy tool to display several dependent variable against one independent variable. However, it is reccomended that no more than 5-10 lines on a single graph, any more than that and it become difficult to interpret.

```python
# compare the trends of top 5 countries that contribute the most to immagration to canada

inplace=True

df_can.sort_values(by='Total', ascending=False, axis=0, inplace=True)

#get the top 5 entries
df_top5= df_can.head(5)
df_top5
```

```python
# we have to tanspose
df_top5=df_top5[years].transpose()

df_top5
```

```python
#plot the dataframe

df_top5.index=df_top5.index.map(int)
df_top5.plot(kind='line', figsize=(14,8))

plt.title('Immigration trends of top 5 countries')
plt.xlabel('Years')
plt.ylabel('No of immigrants')

plt.show()
```
